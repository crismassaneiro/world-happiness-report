Tratando as variável dos datasets para gerar um único dataset

* Datasets com variável Country or region devem ter o nome da coluna alterado para country
* A coluna de region deve ser retirada
* Datasets com variável Happiness.Rank renomear para happiness_rank
* Datasets com variável Happiness.Score renomear para happiness_score
* Váriavel de escore para os anos de 2018 e 2019 devem ter o nome corrigido
* Remover a variável standard error do ano de 2015 
* A variável (GDP per capita) precisa ter o nome normalizado entre todos os datasets
* Para o dataset de 2016, remover as variáveis lower confidence interval, upper confidence interval


Renomeando as colunas dos datasets

2015

```python
print(list(dataset2015.columns))
```

```python
dataset2015.rename(columns={'Country': 'country',
                            'Happiness Rank': 'happiness_rank',
                            'Happiness Score': 'happiness_score',
                            'Economy (GDP per Capita)': 'gdp_per_capita',
                            'Family': 'family',
                            'Health (Life Expectancy)': 'life_expect',
                            'Freedom': 'freedom',
                            'Trust (Government Corruption)': 'trust_gc',
                            'Generosity': 'generosity'},inplace = True)
dataset2015.drop(columns=['Region','Standard Error','Dystopia Residual'],axis=1, inplace=True)                   
```

```python
dataset2016.rename(columns={'Country': 'country',
                            'Happiness Rank': 'happiness_rank',
                            'Happiness Score': 'happiness_score',
                            'Economy (GDP per Capita)': 'gdp_per_capita',
                            'Family': 'family',
                            'Health (Life Expectancy)': 'life_expect',
                            'Freedom': 'freedom',
                            'Trust (Government Corruption)': 'trust_gc',
                            'Generosity': 'generosity'},inplace = True)
dataset2016.drop(columns=['Region','Lower Confidence Interval', 'Upper Confidence Interval', 'Dystopia Residual'],axis=1, inplace=True)  
```
