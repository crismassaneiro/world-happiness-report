Imports

```python
import pandas as pd  # ler e transformar dados 
import numpy as np # algebra linear
import matplotlib.pyplot as plt #gerar gráficos
import seaborn as sns #gerar gráficos
import warnings #logs
warnings.filterwarnings("ignore")
```

Importando CSV

Selecionando as 8 cidades de cada ano com o maior score de felicidade

```python
sort_df_2015 = all_datasets[all_datasets['year'] == 2015].sort_values(by='happiness_score', ascending=False).head(8)
sort_df_2016 = all_datasets[all_datasets['year'] == 2016].sort_values(by='happiness_score', ascending=False).head(8)
sort_df_2017 = all_datasets[all_datasets['year'] == 2017].sort_values(by='happiness_score', ascending=False).head(8)
sort_df_2018 = all_datasets[all_datasets['year'] == 2018].sort_values(by='happiness_score', ascending=False).head(8)
sort_df_2019 = all_datasets[all_datasets['year'] == 2019].sort_values(by='happiness_score', ascending=False).head(8)
```

Selecionando as 8 cidades de cada ano com o menor score de felicidade

```python
bad_df_2015 = all_datasets[all_datasets['year'] == 2015].sort_values(by='happiness_score', ascending=True).head(8)
bad_df_2016 = all_datasets[all_datasets['year'] == 2016].sort_values(by='happiness_score', ascending=True).head(8)
bad_df_2017 = all_datasets[all_datasets['year'] == 2017].sort_values(by='happiness_score', ascending=True).head(8)
bad_df_2018 = all_datasets[all_datasets['year'] == 2018].sort_values(by='happiness_score', ascending=True).head(8)
bad_df_2019 = all_datasets[all_datasets['year'] == 2019].sort_values(by='happiness_score', ascending=True).head(8)
```

```python
f, axs = plt.subplots(5,2, figsize = (14,25), constrained_layout = True)
ax=sns.barplot(x="country", y="happiness_score", data=sort_df_2015 , ax = axs[0, 0],palette="cubehelix")
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="country", y="happiness_score", data=bad_df_2015 , ax = axs[0, 1],palette="cubehelix")
ax.tick_params(axis = 'x', rotation = 45)

ax=sns.barplot(x="country",y="happiness_score", data=sort_df_2016, ax = axs[1, 0],palette="cubehelix")
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="country",y="happiness_score", data=bad_df_2016, ax = axs[1, 1],palette="cubehelix")
ax.tick_params(axis = 'x', rotation = 45)


ax=sns.barplot(x="country",y="happiness_score", data=sort_df_2017, ax = axs[2, 0],palette="cubehelix")
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="country",y="happiness_score", data=bad_df_2017, ax = axs[2, 1],palette="cubehelix")
ax.tick_params(axis = 'x', rotation = 45)

ax=sns.barplot(x="country",y="happiness_score", data=sort_df_2018, ax = axs[3, 0],palette="cubehelix")
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="country",y="happiness_score", data=bad_df_2018, ax = axs[3, 1],palette="cubehelix")
ax.tick_params(axis = 'x', rotation = 45)

ax=sns.barplot(x="country",y="happiness_score", data =sort_df_2019, ax = axs[4, 0],palette="cubehelix")
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="country",y="happiness_score", data=bad_df_2019, ax = axs[4, 1],palette="cubehelix")
ax.tick_params(axis = 'x', rotation = 45)


axs[0, 0].title.set_text('8 países mais felizes em 2015')
axs[0, 1].title.set_text('8 países menos felizes em 2015')

axs[1, 0].title.set_text('8 países mais felizes em 2016')
axs[1, 1].title.set_text('8 países menos felizes em 2016')

axs[2, 0].title.set_text('8 países mais felizes em 2017')
axs[2, 1].title.set_text('8 países menos felizes em 2017')

axs[3, 0].title.set_text('8 países mais felizes em 2018')
axs[3, 1].title.set_text('8 países menos felizes em 2018')

axs[4, 0].title.set_text('8 países mais felizes em 2019')
axs[4, 1].title.set_text('8 países menos felizes em 2019')



```

Análise para comparar a diferença de score com o passar do anos

```python
df_2015 = all_datasets[all_datasets["year"] == 2015]
df_2016 = all_datasets[all_datasets["year"] == 2016]
df_2017 = all_datasets[all_datasets["year"] == 2017]
df_2018 = all_datasets[all_datasets["year"] == 2018]
df_2019 = all_datasets[all_datasets["year"] == 2019]

df=pd.merge(df_2015,df_2016,on="country",suffixes=("_2015","_2016"))
df2=pd.merge(df,df_2017,on="country",suffixes=("_2016","_2017"))
df3=pd.merge(df2,df_2018,on="country",suffixes=("_2017","_2018"))
df4=pd.merge(df3,df_2019,on="country",suffixes=("_2018","_2019"))

df4.rename(columns={'happiness_score': 'happiness_score_2019'},inplace=True)

df_graph = df4[["country","happiness_score_2015","happiness_score_2016","happiness_score_2017","happiness_score_2018","happiness_score_2019"]]

df_graph['score_changes_2015_2016']=df_graph['happiness_score_2015']-df_graph["happiness_score_2016"].round(3)
df_graph['score_changes_2016_2017']=df_graph['happiness_score_2016']-df_graph["happiness_score_2017"].round(3)
df_graph['score_changes_2017_2018']=df_graph['happiness_score_2017']-df_graph["happiness_score_2018"].round(3)
df_graph['score_changes_2018_2019']=df_graph['happiness_score_2018']-df_graph["happiness_score_2019"].round(3)

df_graph['score_changes_2015_2019']=df_graph['happiness_score_2015']-df_graph["happiness_score_2019"].round(3)

df_graph2 = df_graph.sort_values(by = 'score_changes_2015_2019', ascending = False)
df_graph3 = df_graph.sort_values(by = 'score_changes_2015_2016', ascending = False)
df_graph4 = df_graph.sort_values(by = 'score_changes_2016_2017', ascending = False)
df_graph5 = df_graph.sort_values(by = 'score_changes_2017_2018', ascending = False)
df_graph6 = df_graph.sort_values(by = 'score_changes_2018_2019', ascending = False)
```

Análise do aumento/diminuição do score de felicidade durante os anos de 2015/2019

```python
f2, axs2 = plt.subplots(1,2, figsize = (14,6), constrained_layout = True)
ax2=sns.barplot(y = df_graph2["country"][:8], x = df_graph2["score_changes_2015_2019"][:8],palette="inferno", ax=axs2[0])
ax2.tick_params(axis = 'x', rotation = 45)
ax2=sns.barplot(y = df_graph2["country"][-8:], x = df_graph2["score_changes_2015_2019"][-8:],palette="inferno", ax=axs2[1])
ax2.tick_params(axis = 'x', rotation = 45)

axs2[0].title.set_text('Os 8 países com maior diminuição no escore de felicidade de 2015 a 2019')
axs2[1].title.set_text('Os 8 países com maior aumento no escore de felicidade de 2015 a 2019')
```
